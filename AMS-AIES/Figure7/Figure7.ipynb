### Figure 7 Kernel Sensitivity heat map
Input data:
RMSE AND IOA of surface zonal wind stress and OLR. \
The files 'ws_rmse.txt', 'ws_ioa.txt', 'olr_rmse.txt', and 'olr_ioa.txt' are generated by CNN_Kernel_Sensitivty_Test.ipynb

```python
import numpy as np
import pandas as pd
import proplot as pplt
import matplotlib.pyplot as plt
```

```python
pplt.rc.update({'meta.width':1, 'ticklabelweight':'bold', 'axeslabelweight':'bold','titleweight':'bold','fontname':'Noto Sans','fontsize':32,'titlesize':34})
```

```python
kernel1 = np.arange(60, 130, 10).tolist()
kernel2 = np.arange(10, 70, 10).tolist() 
```

#### Load RMSE and IOA for zonal surface wind stress

```python
df1 = pd.read_csv('ws_rmse.txt',header=None,index_col=False)
df2 = pd.read_csv('ws_ioa.txt',header=None,index_col=False)

ex1 = df1.to_numpy().squeeze()
ex2 = df2.to_numpy().squeeze()

rmse_ws = np.ones((6,7))  
ioa_ws = np.ones((6,7))  

m = 0
n = 6
for i in range(7):
    rmse_ws[:,i] = ex1[m:n]
    ioa_ws[:,i] = ex2[m:n]
    m = m + 6
    n = n + 6
```

#### Load RMSE and IOA for OLR

```python
df1 = pd.read_csv('olr_rmse.txt',header=None,index_col=False)
df2 = pd.read_csv('olr_ioa.txt',header=None,index_col=False)

ex1 = df1.to_numpy().squeeze()
ex2 = df2.to_numpy().squeeze()

rmse_olr = np.ones((6,7))  
ioa_olr = np.ones((6,7))  

m = 0
n = 6
for i in range(7):
    rmse_olr[:,i] = ex1[m:n]
    ioa_olr[:,i] = ex2[m:n]
    m = m + 6
    n = n + 6
```

```python
f,ax = pplt.subplots(nrows=1,ncols=2,refwidth=8,refheight=5,share=0)

f.set_dpi(150)

c1 = ax[0].pcolor(rmse_ws*10**3, edgecolors='k', linewidths=4, cmap='Rocket')
cb1 = plt.colorbar(c1, ax=ax[0],aspect=40)
ax[0].set_xticks(np.arange(len(kernel1)));
ax[0].set_xticklabels(kernel1,rotation='horizontal');
ax[0].set_yticks(np.arange(len(kernel2)));
ax[0].set_yticklabels(kernel2,rotation='horizontal');
ax[0].set_ylabel('Kernel 2');
ax[0].set_xlabel('Kernel 1');
ax[0].set_title(r'($\times$ 10$^{-3}$) RMSE $[Nm^{-2}]$');

c2 = ax[1].pcolor(ioa_ws*100, edgecolors='k', linewidths=4, cmap='Rocket_r')
cb2 = plt.colorbar(c2, ax=ax[1],aspect=40)
ax[1].set_xticks(np.arange(len(kernel1)));
ax[1].set_xticklabels(kernel1,rotation='horizontal');
ax[1].set_yticks(np.arange(len(kernel2)));
ax[1].set_yticklabels(kernel2,rotation='horizontal');
ax[1].set_ylabel('Kernel 2');
ax[1].set_xlabel('Kernel 1');
ax[1].set_title('IOA');

ax[0].format(xtickminor=False,ytickminor=False)
ax[1].format(xtickminor=False,ytickminor=False)

ax.format(abc=('a)','b)'),abcloc='l')
#########################################################################################33
f.suptitle('Zonal Wind Stress')


f.savefig('Fig7a.png',dpi=150)
```

```python
f,ax = pplt.subplots(nrows=1,ncols=2,refwidth=8,refheight=5,share=0)

f.set_dpi(150)

c3 = ax[0].pcolor(rmse_olr, edgecolors='k', linewidths=4, cmap='Rocket')
cb3 = plt.colorbar(c3, ax=ax[0],aspect=40)
ax[0].set_xticks(np.arange(len(kernel1)));
ax[0].set_xlabel('Kernel 1');
ax[0].set_xticklabels(kernel1,rotation='horizontal');
ax[0].set_yticks(np.arange(len(kernel2)));
ax[0].set_yticklabels(kernel2,rotation='horizontal');
ax[0].set_ylabel('Kernel 2');
ax[0].set_title(r'RMSE $[Wm^{-2}]$');


c4 = ax[1].pcolor(ioa_olr*100, edgecolors='k', linewidths=4, cmap='Rocket_r')
cb4 = plt.colorbar(c4, ax=ax[1],aspect=40)
ax[1].set_xticks(np.arange(len(kernel1)));
ax[1].set_xlabel('Kernel 1');
ax[1].set_xticklabels(kernel1,rotation='horizontal');
ax[1].set_yticks(np.arange(len(kernel2)));
ax[1].set_yticklabels(kernel2,rotation='horizontal');
ax[1].set_ylabel('Kernel 2');
ax[1].set_title('IOA');

ax[0].format(xtickminor=False,ytickminor=False)
ax[1].format(xtickminor=False,ytickminor=False)

ax.format(abc=('c)','d)'),abcloc='l')
f.suptitle('OLR')
f.savefig('Fig7b.png',dpi=150)
```
